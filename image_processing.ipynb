<h2> Image processing in python using cv2</h2>

```python
# Import modules
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
import pandas as pd
from google.colab.patches import cv2_imshow 
```

```python
# Read the image using cv2 imread function
img =cv2.imread("/img/citi.jpeg") # the original coloured image
```

```python
#to read the image in grayscale
gray_scale_img =cv2.imread("/img/citi.jpeg", 0) # the original coloured image
```

```python
cv2_imshow(gray_scale_img)
```

# Resizing The Image

```python

scale_value = 85 # 85% of the original size // the coloured image
```

```python
width = int(img.shape[1] * scale_value / 90)
height = int(img.shape[0] * scale_value / 80)
dim = (width, height)
resized=cv2.resize(img,dim)
cv2_imshow(resized)
cv2.waitKey(0)
```

```python
#the image is now a numpy array 
print(type(img))
```

```python
#Image shape
h, w, c = img.shape
print("Height:", h, "Width:", w, "Number of Channels:", c)
```

```python
#save img in grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imwrite('citi.jpg', gray)
```

```python
# Construct a Synthintic image from the img
syn = np.zeros([256, 256])
syn[0:32, :] = 55 
syn[:, :32] = 55 
syn[:, 224:256] = 40 
syn[224:, :] = 40 
syn[32:64, 32:224] = 100 
syn[64:224, 32:64] = 100 
syn[64:224, 192:224] = 100 
syn[192:224, 32:224] = 100 
syn[64:96, 64:192] = 160 
syn[96:192, 64:96] = 160 
syn[96:192, 160:192] = 160 
syn[160:192, 64:192] = 160 
syn[96:160, 96:160] = 220
plt.imshow(syn)
```

```python
im_g = cv2.imread("/img/citi.jpeg", 1)
# Loading the image 

half = cv2.resize(im_g, (0, 0), fx = 0.2, fy = 0.2)
bigger = cv2.resize(im_g, (1550, 1610))

stretch_near = cv2.resize(im_g, (480, 740),
			interpolation = cv2.INTER_NEAREST)


title =["First", "Second", "Bigger", "Interpolation"]
shapes =[im_g, half, bigger, stretch_near]
n = 4

for i in range(n):
	plt.subplot(2, 3, i + 1)
	plt.title(title[i])
	plt.imshow(shapes[i])

plt.show()
```
